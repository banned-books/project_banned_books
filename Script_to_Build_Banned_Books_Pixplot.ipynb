# Run the PixPlot Build With Our Code and Our Parameters

Now, run PixPlot! The basic command is `pixplot --images "path/to/images/*.jpg"`. Using the demo materials we downloaded, we can also add metadata using the metadata flag. See the [github repo](https://github.com/YaleDHLab/pix-plot) for all of the parameters that you can control.

````
!pixplot --images "/content/drive/MyDrive/book_cover_images/*" --metadata "/content/drive/MyDrive/Colab Notebooks/data/cleaned_trained_data/metadata.csv" --metric 'correlation' --n_clusters 2 --seed 42
````
